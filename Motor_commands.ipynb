```python
import numpy as np
import matplotlib.pyplot as plt

# create some data for the plot (sine wave)
x = np.linspace(0, 10*np.pi, 500)
y = np.sin(x)

# create a figure and plot the data
fig, ax = plt.subplots()
ax.plot(x, y)

# enable zooming
ax.set(xlim=(0, 10*np.pi), ylim=(-1, 1))
ax.set_title('Interactive plot with zooming')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_autoscale_on(False)
ax.format_coord = lambda x, y: f'x={x:.2f}, y={y:.2f}'

def on_scroll(event):
    ax.set_xlim(ax.get_xlim()[0]-event.step, ax.get_xlim()[1]+event.step)
    ax.figure.canvas.draw()

# connect the on_scroll function to the figure canvas
fig.canvas.mpl_connect('scroll_event', on_scroll)

plt.show()
```
